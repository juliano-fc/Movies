#### Processamanto de Texto

```python
# Esse loop itera sobre os elementos de uma estrutura de dados. 
# Usamos a ast.literal_eval(obj) para avaliar/converter a string obj em uma estrutura de dados python 

def converter(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L 
```

```python
# Aplicando a função na colune de gênero do filme 
df_filmes['genres'] = df_filmes['genres'].apply(converter)
df_filmes.head()
```

```python
# Aplicando a função na colune de keywords 
df_filmes['keywords'] = df_filmes['keywords'].apply(converter)
df_filmes.head()
```

```python
# Vamos seguir aplicando o AST à coluna de 'cast' mas extrair apenas os 4 primeiros elementos, considerando que os principais atores estão nas primeiras posições

def converter2(obj):
    L= []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 4:
            L.append(i['name'])
            counter = counter +1
        else:
            break
    return L 
```

```python
df_filmes['cast'] = df_filmes['cast'].apply(converter2)
df_filmes.head()
```

```python
# Vamos filtrar a coluna 'crew' escolhendo apenas o diretor 

def fetch_director(obj):
    L= []
    # Converte a string 'obj' em um objeto python e itera sobre seus elementos 
    for i in ast.literal_eval(obj):
        # Verifica se o trabalho ('job') do elemento atual é 'Director' 
        if i['job'] == 'Director':
            # Adiciona o nome do diretor (valor da variável 'name') à lista L 
            L.append(i['name'])
            # Interrompe o loop após encontrar o primeiro diretor 
            break
    return L
```

```python
df_filmes['crew'] = df_filmes['crew'].apply(fetch_director)
df_filmes.head()
```

#### Limpeza nos dados 

```python
#Separamos a string da coluna overview por espaço em branco 

df_filmes['overview'] = df_filmes['overview'].apply(lambda x:x.split())
df_filmes.head()
```

```python
# Replace do espaço vazio (remove o espaço vazio)
df_filmes['genres'] = df_filmes['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df_filmes['keywords'] = df_filmes['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df_filmes['cast'] = df_filmes['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df_filmes['crew'] = df_filmes['crew'].apply(lambda x:[i.replace(" ","") for i in x])

df_filmes.head()
```

```python
# Preparação dos dados para vetorização 
# Criamos a coluna 'tags', um vetor de strings com valores das colunas 

df_filmes['tags'] = df_filmes['overview'] + \
                    df_filmes['genres'] + \
                    df_filmes['keywords'] + \
                    df_filmes['cast'] + \
                    df_filmes['crew']

df_filmes.head()
```

```python
df_filmes_novo = df_filmes[['movie_id','title','tags']]
df_filmes_novo.head()
```

```python
# Join das strings para simplificar o vetor 
df_filmes_novo['tags'] = df_filmes_novo['tags'].apply(lambda x:" ".join(x))
# Coloca tudo em minúsculo 
df_filmes_novo['tags'] = df_filmes_novo['tags'].apply(lambda x:x.lower())
df_filmes_novo.head()
```

#### Parse e vetorização 

Adotaremos o processo de Stemming, ou seja, é o processo de redução de uma palavra ao seu radical que está ligado a sufixos e prefixos ou à raízes de palvras conhecidas como 'lemma'. É uma estratégia de tratamento de texto que simplifica o conjunto de dados.

```python
# Função stemming
parser_ps = PorterStemmer ()

def stem(texto):
    y = []
    for i in texto.split():
        y.append(parser_ps.stem(i))
    
    return " ".join(y)
```

```python
# Vamos aplicar a função à coluna de tags 
df_filmes_novo['tags'] = df_filmes_novo['tags'].apply(stem)
df_filmes_novo.head()
```

A verorização no contexo de processamento de linguagem natural (PLN) é o processo de converter o texto em uma representação numérica, geralmente na forma de vetores. Esse processo é fundamental para que algoritmos de aprendizado de máquina possam trabalhar com dados textuais. 

CountVectorizer é uma técnica de vetorização que converte uma coleção de documentos de texto em uma matriz de contagem de tokens, de forma simplificada, ela conta quantas vezes cada palavra ocorrem em cada documento. Neste caso, limitamos a buscar as 500 palavras mais frequentes. As palavras de parada em ingês (stop words) devem ser removidas do texto 

```python
# Cria o vetorizador com, no máximo, 5000 atributos 
from sklearn.feature_extraction.text import CountVectorizer    
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(df_filmes_novo['tags']).toarray()
```

```python
len(cv.get_feature_names_out())
```

```python
vectors
```

#### Cálculo da distânica entre os vetores 

```python
similaridades = cosine_similarity(vectors)
```

#### Construindo o sistema de recomendação 

```python
def sist_recomencao(movie):

    # Obtém o índice do filme passado como argumento (o que o usuário assistiu)
    index = df_filmes_novo[df_filmes_novo['title'] == movie].index[0]

    # Verificando os filmes com vetores com menor distância para o filme passado como arguemento 
    distancias = sorted(list(enumerate(similaridades[index])),reverse=True,key = lambda x:x[1])

    # Então consideramos os 5 filmes com menor distância, ou seja, maior similaridade 
    for i in distancias[1:6]:
        print(df_filmes_novo.iloc[i[0]].title)
```

#### Aplicando o sistema de recomendação 

```python
# Quais as recomendações de filmes para quem assistiu ao filme: Avengers: Age of Ultron? 
sist_recomencao('Avengers: Age of Ultron')
```

```python
# Quais as recomendações de filmes para quem assistiu ao filme: Titanic? 
sist_recomencao('Titanic')
```
